**VISUALIZATION**

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['Feature_1'], data['Feature_2'])
plt.xlabel('Feature_1')
plt.ylabel('Feature_2')
plt.title('Scatter Plot of Feature_1 vs Feature_2')
plt
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=11,max_iter=100, random_state=1)
kmeans
```

```python
kmeans.fit(data)
```

```python
data['Cluster'] = kmeans.predict(data)
```

```python
data.head(5)
```

```python
plt.scatter(data['Feature_1'], data['Feature_2'], c=data['Cluster'], cmap='rainbow')
plt.xlabel('Feature_1')
plt.ylabel('Feature_2')
plt.title('Clusters of Feature_1 and Feature_2')
plt
```
